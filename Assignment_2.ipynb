Question 1: What is hypothesis testing in statistics?

    Answer:Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data.
    It helps us determine whether there is enough evidence to support a claim or reject it.

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
    Answer:Null Hypothesis (H₀)

    Assumes no effect, no difference, or status quo

    Example:

    H₀: The mean weight = 50 kg

    Alternative Hypothesis (H₁ or Hₐ)

    Represents the claim we want to test

    Indicates an effect or difference

    Example:

    H₁: The mean weight ≠ 50 kg

    | Null Hypothesis          | Alternative Hypothesis  |
    | ------------------------ | ----------------------- |
    | No change/effect         | Change/effect exists    |
    | Default assumption       | What we try to prove    |
    | Rejected or not rejected | Accepted if H₀ rejected |


Question 3: Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.

    Answer:The significance level (α) is the probability of rejecting the null hypothesis when it is actually true.

    Common values: 0.05 (5%), 0.01 (1%)

    α = 0.05 means:

    We accept a 5% chance of making a wrong decision

    Role

    If p-value ≤ α → Reject H₀

    If p-value > α → Fail to reject H₀

Question 4: What are Type I and Type II errors? Give examples of each.

    Answer:Type I Error

    Rejecting a true null hypothesis

    Called False Positive

    Probability = α

    Example:
    Saying a medicine works when it actually doesn’t.

    Type II Error

    Failing to reject a false null hypothesis

    Called False Negative

    Probability = β

    Example:
    Saying a medicine doesn’t work when it actually does.


Question 5: What is the difference between a Z-test and a T-test? Explain when to use
each.

    Answer: | Feature             | Z-test                 | T-test                |
    | ------------------- | ---------------------- | --------------------- |
    | Sample size         | Large (n ≥ 30)         | Small (n < 30)        |
    | Population variance | Known                  | Unknown               |
    | Distribution        | Normal (Z)             | Student’s t           |
    | Accuracy            | Higher with large data | Better for small data |

```python
# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
# Python. Show the Python code and interpret the results.
# sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
# 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
# 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
# 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

import numpy as np
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

mu = 50  # hypothesized mean

z_stat = (sample_mean - mu) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Z-statistic:", z_stat)
print("P-value:", p_value)
```

```python
# Question 8: Write a Python script to simulate data from a normal distribution and
# calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Simulate data
data = np.random.normal(loc=100, scale=15, size=1000)

mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

# 95% Confidence Interval
z = norm.ppf(0.975)
margin = z * (std / np.sqrt(n))
ci = (mean - margin, mean + margin)

print("95% Confidence Interval:", ci)

# Plot
plt.hist(data, bins=30)
plt.axvline(ci[0])
plt.axvline(ci[1])
plt.title("Normal Distribution with 95% Confidence Interval")
plt.show()
```

```python
# Question 9: Write a Python function to calculate the Z-scores from a dataset and
# visualize the standardized data using a histogram. Explain what the Z-scores represent
# in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

data = np.random.normal(50, 10, 1000)
z_scores = calculate_z_scores(data)

plt.hist(z_scores, bins=30)
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Histogram of Z-scores")
plt.show()
```
